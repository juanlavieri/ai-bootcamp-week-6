```python
# Dependencies
import requests
import pandas as pd
```

```python
# list of TV show titles to query
tv_shows = ["Altered Carbon", "Grey's Anatomy", "This is Us", 
            "Stranger Things", "Family Law", "Good Omens",
            "Vikings", "Shameless", "Arrow", "Dirk Gently"]

# TV Maze show search base URL
base_url = "http://api.tvmaze.com/search/shows?q="

# set up lists to hold response data for name and rating
titles = []
ratings = []
all_results = []

for show in tv_shows:
    response = requests.get(f"http://api.tvmaze.com/search/shows?q={show}")
    data = response.json()

    if data:
        first_show = data[0]['show']
        titles.append(first_show['name'])
        ratings.append(first_show.get('rating', {}).get('average'))
        all_results.append(first_show)
```
